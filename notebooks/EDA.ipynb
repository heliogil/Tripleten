# EDA - Vehicle Ads Dashboard

Análise exploratória básica do conjunto de dados `vehicles_us.csv`.

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv')
car_data.head()
```

```python
fig = px.histogram(car_data, x='odometer', nbins=50, title='Distribuição da quilometragem')
fig.show()
```

```python
fig = px.scatter(car_data, x='odometer', y='price', color='condition', title='Preço vs quilometragem')
fig.show()
```
